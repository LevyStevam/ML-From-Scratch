## Teste da regress√£o linear GD

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from models.linear_regression.linear_regression_gd import LinearRegressionGD
```

```python
df = pd.read_csv('../data/processed/artificial1d.csv', delimiter=",")

X, y = df.iloc[:, 0], df.iloc[:, 1]
```

```python
linear_regression_gd = LinearRegressionGD()
linear_regression_gd.fit(X, y, epochs=1000, learning_rate=0.01, plot_learning_curve=True)
```

```python
X_test = np.random.uniform(-1, 1, 1000)
```

```python
predictions = linear_regression_gd.predict(X_test)
```

```python
plt.scatter(X, y, color='blue', label='Dados', alpha=0.6)

plt.plot(X_test, predictions, color='red', label='Reta de regress√£o')

plt.xlabel('Altura')
plt.ylabel('Comprimento do cateter')
plt.legend()
plt.show()
```
