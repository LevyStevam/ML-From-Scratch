## Algoritmo K-Means

O **K-Means** é um algoritmo de agrupamento não supervisionado que particiona os dados em \(K\) clusters, onde cada ponto pertence ao cluster com o centróide mais próximo. É amplamente utilizado para descobrir padrões e estruturas ocultas em conjuntos de dados.

### Intuição
O objetivo do K-Means é minimizar a soma das distâncias quadradas entre cada ponto e o centróide do cluster ao qual ele pertence. Isso é feito iterativamente ajustando os centróides e reatribuindo os pontos aos clusters.

### Fórmula do Erro de Quantização
O objetivo do K-Means é minimizar a seguinte função:

\[
J(C) = \sum_{k=1}^{K} \sum_{x_i \in C_k} ||x_i - m_k||^2
\]

Onde:
- \(C_k\): Cluster \(k\).
- \(m_k\): Centróide do cluster \(k\).
- \(||x_i - m_k||^2\): Distância Euclidiana quadrada entre o ponto \(x_i\) e o centróide \(m_k\).
