```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.utils import shuffle

from models.linear_regression_gd import LinearRegressionGD
```

```python
data = np.loadtxt('../data/processed/artificial1d.csv', delimiter=",")

X = data[:,0:1]
y = data[:,1]

X, y = shuffle(X, y, random_state=10)

X_train = data[:,0:1]
y_train = data[:,1]
X_test = data[:,0:1]
y_test = data[:,1]
```

```python
linear_regression_gd = LinearRegressionGD(w0=0.0, w1=0.0)
linear_regression_gd.fit(X_train, y_train, epochs=200, learning_rate=0.1)
```

```python
linear_regression_gd.w0
```

```python
predictions = linear_regression_gd.predict(X_test)
```

```python
plt.scatter(X_test, y_test, color='blue', label='Dados', alpha=0.6)

plt.plot(X_test, predictions, color='red', label='Reta de regress√£o')

plt.xlabel('Altura')
plt.ylabel('Comprimento do cateter')
plt.legend()
plt.show()
```
