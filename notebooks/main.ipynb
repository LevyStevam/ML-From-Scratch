```python
import numpy as np
import matplotlib.pyplot as plt

from models.linear_regression import LinearRegression

root = '../'
```

```python
data = np.loadtxt(root + 'data/processed/artificial1d.csv', delimiter=",")
X = data[:,0:1]
y = data[:,1]
```

```python
linear_regression = LinearRegression(w0=0.0, w1=0.0)
linear_regression.fit(X, y, epochs=200, learning_rate=0.2)
```

```python
linear_regression.w0
```

```python
predictions = linear_regression.predict(X)
```

```python
plt.scatter(X, y, color='blue', label='Dados', alpha=0.6)

plt.plot(X, predictions, color='red', label='Reta de regress√£o')

plt.xlabel('Altura')
plt.ylabel('Comprimento do cateter')
plt.legend()
plt.show()
```
