## Teste da regress√£o linear GD

```python
import numpy as np
import matplotlib.pyplot as plt

from models.linear_regression.linear_regression_gd import LinearRegressionGD
```

```python
data = np.loadtxt('../data/processed/artificial1d.csv', delimiter=",")

X = data[:,0:1]
y = data[:,1]
```

```python
linear_regression_gd = LinearRegressionGD(w0=0.0, w1=0.0)
linear_regression_gd.fit(X, y, epochs=250, learning_rate=0.1, plot_learning_curve=True)
```

```python
predictions = linear_regression_gd.predict(X)
```

```python
plt.scatter(X, y, color='blue', label='Dados', alpha=0.6)

plt.plot(X, predictions, color='red', label='Reta de regress√£o')

plt.xlabel('Altura')
plt.ylabel('Comprimento do cateter')
plt.legend()
plt.show()
```
